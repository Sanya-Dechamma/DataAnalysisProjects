```python
#Graph of product category at each store location
```

```python
grouped = df.groupby(['store_location', 'product_category']).size().reset_index(name='count')

pivot_df = grouped.pivot_table(index='store_location', columns='product_category', values='count', fill_value=0)

# Plotting
pivot_df.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Product Categories by Store Location')
plt.xlabel('Store Location')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Product Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
grouped = df.groupby(['store_location', 'product_category']).size().reset_index(name='count')

pivot_df = grouped.pivot_table(index='product_category', columns='store_location', values='count', fill_value=0)

# Plotting
pivot_df.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Product Categories by Store Location')
plt.ylabel('Store Location Count')
plt.xlabel('Product Category')
plt.xticks(rotation=45)
plt.legend(title='Store Location', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
df['transaction_time'] = df['transaction_time'].astype(str)
df['transaction_time'] = pd.to_timedelta(df['transaction_time'])
```

```python
df.info()
```

```python
df['transaction_date'] = pd.to_datetime(df['transaction_date'], format='%m/%d/%y', errors='coerce')
```

```python
df.dtypes
```

```python
df.head(5)
```

```python
df['sales'] = df['transaction_qty'] * df['unit_price']
df['datetime'] = df['transaction_date'] + df['transaction_time']
```

```python
df.head(5)
```

```python
df.dtypes
```

```python
#store locations with number of sales
```

```python
df_location = df.groupby('store_location').agg({
    'sales': 'sum',
    'transaction_id': 'count'
})
df_location
```

```python
daily_sales_by_loc = df.groupby(['transaction_date', 'store_location'])['sales'].sum().unstack()
daily_sales_by_loc
```

```python
daily_sales_by_loc.plot(figsize=(14,8), title='Daily sales by store location')

plt.xlabel('Date')
plt.ylabel('Total sales in $')
plt.legend(title='Store location')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df_location = df_location.reset_index()

plt.figure(figsize=(10, 7))
plt.pie(df_location['sales'], labels=df_location['store_location'], autopct='%1.1f%%', startangle=140)
plt.title('Sales Sum by Store Location')
plt.axis('equal')  
plt.show()
```

```python
plt.figure(figsize=(14,8))

plt.stackplot(daily_sales_by_loc.index, daily_sales_by_loc.T, labels=daily_sales_by_loc.columns)

plt.title=('Daily sales by store location(Stacked Area Chart)')
plt.xlabel('Date')
plt.ylabel('Total sales in $')
plt.legend(title='Store location', loc='upper left')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
daily_sales = df.groupby('transaction_date')['sales'].sum().reset_index()
df['day_of_week'] = df['datetime'].dt.day_name()
weekly_sales = df.groupby('day_of_week')['sales'].sum().reindex(
    ['Monday', 'Tueday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']).reset_index()
df['month'] = df['datetime'].dt.to_period('M')
monthly_Sales = df.groupby('month')['sales'].sum().reset_index()
```

```python
weekly_sales = df.groupby(['day_of_week', 'store_location'])['sales'].sum().unstack()

days_order = ['Monday', 'Tueday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekly_sales = weekly_sales.reindex(days_order)
weekly_sales
                    
```

```python
weekly_sales[["Astoria", "Hell's Kitchen", "Lower Manhattan"]].plot(kind="bar", figsize=(14,8))

plt.title('Weekly sales by store location')
plt.xlabel('Day of the week', fontweight='bold')
plt.ylabel('Total sales in $', fontweight='bold')
plt.legend(title='Store location', loc='upper left')

plt.tight_layout()
plt.show()
```
